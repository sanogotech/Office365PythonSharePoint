## Missing Values

```python
# Find the missing values in the dataset

missing_val_features = [feature for feature in dataset.columns if dataset[feature].isnull().sum() > 0]

for feature in missing_val_features:
    print(feature, np.round(dataset[feature].isnull().mean() * 100, 2), '% missing values')
```

```python
# We can also use a bar plot to indicate the columns with missing values

sns.set_style("whitegrid")

missing = dataset.isnull().sum()
missing = missing[missing > 0]
missing.sort_values(inplace=True)

missing.plot.bar()
```

```python
# We can also use a heatmap to visualise the missing values

sns.heatmap(dataset.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

#### Before simply dropping the missing values, see if there is a relationship between missing values and the Sales Price. Since we only have the target values for the training dataset, we will use the train_df to see the relationship.

```python
# Find the relationship between each column with a missing value and the response variable

for feature in missing_val_features:
    data = train_df.copy()
    
    # Make a variable that indicates 1 if the observation is missing or zero otherwise
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    # Calculate the mean SalePrice where the information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar()
    
    plt.xlabel(feature)
    plt.ylabel('Median Sale Price')
    plt.title('House Price vs {}'.format(feature))
    plt.show()
```

- We can see that there is some dependence of the output variable on the missing values.<br>
- We can also observe that some features have missing values only in the test dataset, as using the train dataset we are only getting a single bar. Therefore, it makes sense to use an imputer which can be fitted on the train set and then used to transform the test set.

- Also, it's worth noting that in this dataset, not all NaN values for categorical variables represent the absence of data. For example, for Alley column, NaN values mean No Alley.<br>
This information is given in data_description.txt.<br>
Hence, for such missing values, we should replace them with None, and for others we will use a missing value imputation method.

### Temporal Variables (e.g. Datetime Variables)

The dataset has 4 year variables. We extract information, such as number of years or days, from the datetime variables. One example in this case can be difference in years between the year the house was built and the year the house was sold. (This analysis will be performed in the feature engineering section).

```python
# Find the temporal variables

temporal_feat = [feature for feature in numerical_var if 'Year' in feature or 'Yr' in feature]
```

```python
temporal_feat
```

```python
# Check whether there is a relationship between the Year Sold and the target variable

train_df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price vs Year Sold')
plt.show()
```

**Here we can see that the median house prices are decreasing with the year sold, which does not seem right! Therefore, we explore this variable further.**

```python
# Compare the difference between all year features and year sold

for feature in temporal_feat:
    if feature != 'YrSold':
        data = train_df.copy()
        
        # Capture the difference between the year variable and the year the house was sold
        data[feature] = data['YrSold'] - data[feature]
        
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.title('House Price vs {}'.format(feature))
        plt.show()
```

We can now observe that subtracting the 'Year Sold' feature values from other temporal features gives a good signal, for example, more is the difference between the Year Built and Year Sold, lesser is the Sale Price (and this is what we would expect).

#### Distribution of Continuous Numerical Variables

```python
# Analyse the continuous variables by creating histograms to understand the distribution

for feature in continuous_num_var:
    data = train_df.copy()
    sns.histplot(data, x=data[feature],  bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

**We can see that for some of the continuous variables we have Gaussian distribution and for some we do not. We need to log-transform the skewed data.**

For skewed data, the tail region will act as an outlier for the model and we know that outliers adversely affect the model performance especially regression-based models. There are some statistical models that are robust to outliers such as Tree-based models but skewness will limit the possibility to try other models. Therefore, it is important to transform the skewed data to close enough to a Gaussian or normal distribution.

A log transformation can help to fit a very skewed distribution into a Gaussian distribution. After log transformation, we can easily see the pattern in our data.

### Cardinality of Categorical Variables
(how many unique categories each categorical variable is having)

A categorical variable is said to possess high cardinality if there are too many unique values. One-hot encoding becomes a problem in this case because we need to have a separate column for each unique value. This leads to 2 problems:
- Space consumption
- Curse of dimensionality

Curse of dimensionality is an issue because as the number of features grows, the amount of data we need to accurately distinguish between these features (in order to give us a prediction) and generalise our model (learned function), also grows exponentially.

```python
for feature in cat_var:
    print('{} has {} unique categories'.format(feature, len(dataset[feature].unique()) ))
```

## Spearman Correlation

**Spearman correlation will help us pick up relationships between variables even when they are non-linear**<br>

Spearman correlation is a non-parametric measure of rank correlation (statistical dependence between the rankings of two variables). It assesses how well the relationship between two variables can be described in a monotonic function.<br>

While Pearson assesses linear relationships, Spearman assesses monotonic relationships (whether linear or not).

```python
# First we target-encode our categorical variables

def target_encode(df):
    data = df.copy()
    
    for feature in cat_var:
        labels_ordered = data.groupby(feature)['SalePrice'].mean().sort_values().index
        labels_ordered = {j:i for i, j in enumerate(labels_ordered)}
        data[feature] = data[feature].map(labels_ordered)
        
    return data
        
```

```python
train_encoded = target_encode(train_df)
```

```python
# Spearman correlation plot (heat map) for numerical features
plt.figure(1)
corr = train_encoded[numerical_var].corr()
sns.heatmap(corr)

# Spearman correlation plot for categorical features
plt.figure(2)
corr = train_encoded[cat_var+['SalePrice']].corr()
sns.heatmap(corr)

# Spearman correlation plot for numerical and categorical features
plt.figure(3)
corr = pd.DataFrame(np.zeros([len(numerical_var), len(cat_var)+1]), 
                    index=numerical_var, columns=cat_var+['SalePrice'])

for i in numerical_var:
    for j in cat_var+['SalePrice']:
        corr.loc[i, j] = train_encoded[i].corr(train_encoded[j])
        
sns.heatmap(corr)
```

It is difficult to see from these plots which features are highly correlated. Later, we can use a filter-based feature selection method with Spearman correlation coefficient as our statistical measure to evaluate the relationship between each input variable and the target.<br>
(We want to keep the features that are correlated with the target as well as remove the features which are highly correlated with each other).
